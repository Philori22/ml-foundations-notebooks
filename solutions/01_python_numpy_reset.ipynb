## 9. Verbal explanation drill

Practice saying these out loud without notes:

### A. What is standardisation?
A transform that rescales each feature to have mean near zero and standard deviation near one, usually to help optimisation and make features numerically comparable.

### B. Why do train/test split?
To evaluate whether the model generalises to unseen data rather than only memorising the training set.

### C. Why print shapes?
Because many ML bugs come from mismatched dimensions, and checking shapes is a fast way to verify assumptions before debugging more complex issues.

### D. What does learning mean in ML?
A model has parameters, and training adjusts those parameters to reduce a loss on data, ideally in a way that generalises to unseen examples.
